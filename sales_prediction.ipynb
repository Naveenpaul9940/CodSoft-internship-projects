```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv("/content/advertising.csv")
```

```python
#visualization
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
sns.scatterplot(x='TV',y='Sales',data = data,color = 'red',ax = axes[0])
sns.scatterplot(x='Radio',y='Sales',data = data,color = 'red',ax = axes[1])
sns.scatterplot(x='Newspaper',y='Sales',data = data,color = 'red',ax = axes[2])
plt.show()
```

```python
features = ['TV','Radio','Newspaper']
x= data[features]
y=data['Sales']
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
training_prediction = model.predict(x_train)
print(training_prediction)
```

```python
testing_prediction = model.predict(x_test)
print(testing_prediction)
```

```python
coefficient = model.score(x_test,y_test)
coefficient

```
