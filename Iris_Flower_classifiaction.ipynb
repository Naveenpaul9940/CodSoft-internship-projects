```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv("/content/IRIS.csv")
```

```python
data.head()
```

```python
data.tail()
```

```python
labelencoder = LabelEncoder()
data['species'] = labelencoder.fit_transform(data['species'])
data
```

```python
data.isnull().any()
```

```python
data.info
```

```python
plt.figure(figsize=(8,6));
sns.scatterplot(x=data.sepal_length,y=data.sepal_width,hue=data.species)
plt.title("Sepal length and Sepal width distribution of three flowers")
plt.show()
```

```python
plt.figure(figsize=(8,6));
sns.scatterplot(x=data.petal_length,y=data.petal_width,hue=data.species)
plt.title("Sepal length and Sepal width distribution of three flowers")
plt.show()
```

```python
x = data.drop(columns = 'species',axis = 1)
y = data['species']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
model = DecisionTreeClassifier()
model.fit(x_train,y_train)
```

```python
training_data_prediction = model.predict(x_train)
accuracy_on_training_data = accuracy_score(training_data_prediction,y_train)
print(accuracy_on_training_data)
```

```python
testing_data_prediction = model.predict(x_test)
accuracy_on_testing_data = accuracy_score(testing_data_prediction,y_test)
print(accuracy_on_testing_data)
```

```python
input = [5.1,3.5,1.4,0.2]
input_as_numpy_arrays = np.asarray(input)
input_data_reshaped = input_as_numpy_arrays.reshape(1,-1)
prediction = model.predict(input_data_reshaped)

if prediction ==0:
  print("Iris-setosa")
elif prediction == 1:
  print("Iris-versicolor")
else:
  print("Iris-virginicia")
```
